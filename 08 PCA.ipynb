# PCA Implementation

```python
## Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
## PCA Components matrix or covariance Matrix
pca.components_
```

```python
PCA_dataset=pd.DataFrame(data=wine_pca,columns=['component1', 'component2', 'component3', 'component4', 
                                                                  'component5', 'component6','component7', 'component8', 'component9',
                                                                 'component10', 'component11', 'component12', 'component13'] )
PCA_dataset.head()
```

```python
## The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
plt.bar(range(1, len(var)+1),var)
```

# From above bar chart it shows that from all these 13 parameters first bar gives the most variance followed by others.

```python
## Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
## Variance plot for PCA components obtained 
plt.plot(var1,color='red')
```

# The line graph will show an increasing trend as we move from the first principal component to the last one. This indicates that each additional principal component contributes to explaining more variance in the data.
# Secondly there is no such knee point above thus it signifies that each factor is contributing to the variance in thid dataset.

```python
principal_component1 = PCA_dataset['component1']
principal_component2 = PCA_dataset['component2']
principal_component3 = PCA_dataset['component3']
```

```python
pca_wine = pd.concat([principal_component1, principal_component2, principal_component3], axis = 1)
pca_wine.head()
```

```python
ax = plt.figure(figsize=(10,10)).gca(projection='3d')
plt.title('3D Principal Component Analysis (PCA)')
ax.scatter(
    xs=principal_component1, 
    ys=principal_component2, 
    zs=principal_component3, 
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
plt.show()
```

# Checking with other Clustering Algorithms

# 1. HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
## As we already have normalized data, we can create Dendrograms
plt.figure(figsize=(25,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
## Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
## Adding clusters to dataset
wine3=wine.copy()
wine3['clustersid']=hclusters.labels_
wine3
```

# cluster is assigned to the dataset using hierarchical clustering.

# 2. K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

## As we already have normalized data
## We can use Elbow Graph to find optimum number of  clusters (K value) from K values range
## The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
## random state can be anything from 0 to 42

```python
## within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
## Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=3

```python
## Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
## Assign clusters to the data set
wine4=wine.copy()
wine4['clusters3id']=clusters3.labels_
wine4
```

# cluster is assigned to the dataset using K means clustering.
